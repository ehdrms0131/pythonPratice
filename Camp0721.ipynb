##예제 6-1 시리즈의 원소에 apply 적용

```python
import seaborn as sns

# titanic 데이터셋에서 age, fare 2개 열을 선택해 데이터프레임 생성
titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','fare']]
df['ten']=10
print(df.head())
```

```python
# 함수 정의
def add_10(n):
  return n + 10

def add_two_obj(a,b):
  return a + b

print(add_10(10))
print(add_two_obj(10,10))
```

##예제 6-2 데이터프레임 원소에 applymap 적용

```python
df_map = df.applymap(add_10)
print(df_map.head())
```

## 예제 6-3 데이터프레임에 apply(axis=0) 적용

```python
def missing_value(series):
  return series.isnull()

res = df.apply(missing_value,axis=0)
print(res.head())
print('\n')
print(type(res))
```

## 예제 6-4 데이터프레임에 apply(axis=0) 적용 2

```python
def min_max(x):
  return x.max() - x.min()

res = df.apply(min_max)
print(res)
print('\n')
print(type(res))
```

##예제 6-5 데이터프레임에 apply() 적용

```python
df['add'] = df.apply(lambda x: add_two_obj(x['age'],x['ten']),axis=1)   
print(df.head())
```

##예제 6-6 데이터프레임에 pipe() 적용

```python
def missing_value(x):
  return x.isnull()

def missing_count(x):
  return missing_value(x).sum()

def total_number_missing(x):
  return missing_count(x).sum

res = df.pipe(missing_value)
print(res.head())
print(type(res))
```
