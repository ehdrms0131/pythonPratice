## 예제 1-15 행,열 변환

```python
examData = {'이름':['서준','우현','인아'],
          '수학':[90,80,70],
          '영어':[98,89,95],
          '음악':[85,95,100],
          '체육':[100,90,90]}
df = pd.DataFrame(examData)
print(df)
print('\n')

# 데이터 프레임 전치(메소드)
df = df.transpose()
print(df)
print('\n')

# 데이터 프레임 전치(클래스)
df = df.T
print(df)
```

##예제 1-16 특정 열 행 인덱스로 설정

```python
examData = {'이름':['서준','우현','인아'],
          '수학':[90,80,70],
          '영어':[98,89,95],
          '음악':[85,95,100],
          '체육':[100,90,90]}
df = pd.DataFrame(examData)
print(df)
print('\n')

# 특정 열을 행 인덱스로 설정
ndf = df.set_index(['이름'])
print(ndf)
print('\n')

ndf2 = df.set_index('음악')
print(ndf2)
print('\n')

ndf3 = ndf.set_index(['수학','음악'])
print(ndf3)
```

##예제 1-17 행 인덱스 재배열

```python
dicData = {'c0':[1,2,3],'c1':[4,5,6],'c2':[7,8,9],'c3':[10,11,12],'c4':[13,14,15]}

# 딕셔너리를 데이터프레임으로 변환 후 지정
df = pd.DataFrame(dicData,index=['r0','r1','r2'])
print(df)
print('\n')

# 인덱스 재지정
newIndex = ['r0','r1','r2','r3','r4']
ndf = df.reindex(newIndex)
print(ndf)
print('\n')

# 인덱스 재지정
newIndex = ['r0','r1','r2','r3','r4']
ndf2 = df.reindex(newIndex,fill_value=0)
print(ndf2)
print('\n')
```

##예제 1-19 데이터프레임 정렬
